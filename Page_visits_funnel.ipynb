```python
import codecademylib
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
#1
print(visits.head())
print(cart.head())
print(checkout.head())
print(purchase.head())

#2
visits_cart = pd.merge(visits, cart, how='left')

#3
print(len(visits_cart))

#4
visits_cart_isnull = len(visits_cart[visits_cart.cart_time.isnull()])
print(visits_cart_isnull)

#5
visits_cart_isnotnull = len(visits_cart[visits_cart.cart_time.notnull()])

percentage_notcart = (1.0* visits_cart_isnull/(visits_cart_isnull + visits_cart_isnotnull))

#6
cart_checkout = pd.merge(cart, checkout, how = 'left')
checkout_null = len(cart_checkout[cart_checkout.checkout_time.isnull()])
print(checkout_null)
checkout_notnull = len(cart_checkout[cart_checkout.checkout_time.notnull()])
print(checkout_notnull)
percentage_notcheckout = (1.0*checkout_null / (checkout_notnull + checkout_null))

#7
all_data = visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
print(all_data.head())

#8
checkout_purchase = pd.merge(checkout, purchase, how='left')
purchase_null = len(checkout_purchase[checkout_purchase.purchase_time.isnull()])
print(purchase_null)
total = len(checkout_purchase)
percentage_notpurchase = (1.0* purchase_null / total)
print(percentage_notpurchase)

#9 
#cart has the highest percentage
print(percentage_notcart)
print(percentage_notcheckout)
print(percentage_notpurchase)

#10
all_data['time_to_purchase'] = \
    all_data.purchase_time - \
    all_data.visit_time

#11
print(all_data.time_to_purchase)

#12
average_timepurchase = all_data.time_to_purchase.mean()
print(average_timepurchase)



```
